```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('NAnderson2020MendeleyMangoNIRData.csv')
df.head()
```

```python
df.describe()
```

```python
df['Cultivar'].unique()
```

```python
# Eliminar columnas que no brindan información
valores_unicos_n = df.nunique()
columnas_sin_info = valores_unicos_n[valores_unicos_n==1].index.tolist()
columnas_sin_info
```

```python
df_clean=df.drop(columnas_sin_info, axis=1)
```

```python
df_clean.describe()
```

```python
df['Cultivar'].unique()
```

```python
df_clean = df_clean[df_clean['Cultivar']=='Caly']
df_clean.head()
```

```python
df_clean['Set'].unique()
```

```python
df_test = df_clean[df_clean['Set']=='Val Ext']
df_test.head()
```

```python
df_cal = df_clean[df_clean['Set']!='Val Ext']
df_cal.head()
```

```python
y_test = df_test['DM']
y_cal = df_cal['DM']

X_test = df_test.iloc[:,9:]
X_cal = df_cal.iloc[:,9:]
```

```python
X_test.head()
```

```python
X_cal.head()
```

```python
print(X_cal.shape)
print(X_test.shape)
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.cross_decomposition import PLSRegression

```

```python
PCR_Pipeline = Pipeline([
    ('scaler',StandardScaler()),
    ('pca',PCA()),
    ('regresion',LinearRegression())
])

MLR_Pipeline = Pipeline([
    ('scaler',StandardScaler()),
    ('regresion',LinearRegression())
])

PLS_Pipeline = Pipeline([
    ('scaler',StandardScaler()),
    ('regresion',PLSRegression())
])
```

```python
# Entrenamiento RLM
MLR_Pipeline.fit(X_cal,y_cal)
```

```python
# Métricas regresión lineal núltiple
pred_mlr = MLR_Pipeline.predict(X_test)
print(r2_score(y_test,pred_mlr))
print(np.sqrt(mean_squared_error(y_test,pred_mlr)))
```

```python
# Entranamineto PCR
PCR_Pipeline.fit(X_cal,y_cal)
```

```python
# Metricas PCR
pred_pcr = PCR_Pipeline.predict(X_test)
print(r2_score(y_test,pred_pcr))
print(np.sqrt(mean_squared_error(y_test,pred_pcr)))
```

```python
pca = PCR_Pipeline.named_steps['pca']
pca.components_
```

```python
explained_variance = pd.DataFrame(pca.explained_variance_ratio_)
explained_variance
```

```python
cumsum = explained_variance.cumsum()
cumsum.plot()
```

### Búsqueda por regilla (elección del número adecuado de componenetes para pca)

```python
from sklearn.model_selection import GridSearchCV, KFold

pca_params = {
    'pca__n_components':np.arange(1,280)
}

cv = KFold(n_splits=10, shuffle=True)
pca_grid = GridSearchCV(PCR_Pipeline,pca_params,cv=cv,scoring='neg_mean_squared_error')

pca_grid.fit(X_cal,y_cal)
```

```python
pca_grid.best_params_['pca__n_components']
```

```python
pca_params = {
    'regresion__n_components':[32,64,100,150,200,250,280]
}

cv = KFold(n_splits=10, shuffle=True)
pca_grid = GridSearchCV(PLS_Pipeline,pca_params,cv=cv,scoring='neg_mean_squared_error')

pca_grid.fit(X_cal,y_cal)
```

```python
pca_grid.best_params_['regresion__n_components']
```

```python
PLS_Pipeline = Pipeline([
    ('scaler',StandardScaler()),
    ('regresion',PLSRegression(n_components=150))
])

PLS_Pipeline.fit(X_cal,y_cal)
```

```python
pred_pls = PLS_Pipeline.predict(X_test)
print(r2_score(y_test,pred_pls))
print(np.sqrt(mean_squared_error(y_test,pred_pls)))
```

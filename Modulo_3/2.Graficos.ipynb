```python
experiencia = [1,3,5,7,9,13]
salario_data_scientists = [11000,15000,18000,25000,35000,55000]
salario_ing_software = [10000,15000,20000,24000,34000,50000]
```

```python
plt.plot(experiencia,salario_data_scientists)
plt.plot(experiencia,salario_ing_software)
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
plt.legend(['Data scientist','Ing Software'])
```

```python
plt.plot(experiencia,salario_data_scientists,
         color = 'r',label='Data Scientist',linestyle='--')
plt.plot(experiencia,salario_ing_software,
         color = 'g',label='Ing Software',linestyle = ':')
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
plt.legend()
plt.grid(color = 'r',linestyle='-',linewidth='0.1')
```

```python
x = ['A','B','C','D']
y = [3,8,1,10]
plt.bar(x,y)
plt.plot(x,y,color='r')
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
```

```python
import numpy as np
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
plt.scatter(x,y,color='g',alpha=0.5)
plt.title('Grafico de dispersin')
```

```python
datos = np.random.randn(10000)

plt.hist(datos,bins=100)
plt.title('Histograma')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
datos = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(datos,vert=True)
plt.show()
```

```python
import pandas as pd

data_range = pd.date_range(start='2023-01-01',end='2023-12-31',freq='D')
precios = np.cumsum(np.random.randn(len(data_range)))+100
```

```python
data = pd.DataFrame({'Date':data_range,'Precios':precios})
```

```python
plt.plot(data['Date'],data['Precios'],label = 'Precio')
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

categorias = ['a','b','c','d']
valores = [5,7,3,8]

datos = [np.random.normal(0,std,100) for std in range(1,4)]

d1 = np.random.rand(100)
d2 = np.random.rand(100)
```

```python
fig, axs = plt.subplots(2,2, figsize=(15,10))
axs[0,0].plot(x,y1, color='r', label = 'Seno')
axs[0,0].plot(x,y2, color='g', label = 'Coseno')
axs[0,0].set_title('Grfico de lineas')
axs[0,0].legend(loc= 'upper right')
axs[0,0].grid(True)


axs[0,1].bar(categorias,valores)
axs[0,1].set_title('Grfico de barras')

axs[1,0].boxplot(datos,vert=True)
axs[1,0].set_title('Grfico de cajas y bigotes')

axs[1,1].scatter(d1,d2)
plt.scatter(d1,d2,color='g',alpha=0.8)
axs[1,1].set_title('Grfico de dispersin')

fig.suptitle('Composicin')
plt.show()
```

En muchas aplicaciones de Ingeniería, por otra parte, más importante que la organización de los datos es la capacidad de hacer muchas operaciones a la vez sobre grandes conjuntos de datos numéricos de manera eficiente. La estructura de datos que sirve para almacenar estas grandes secuencias de números (generalmente de tipo float) es el arreglo.

### Igualdad de matrices

Dos matrices $\textbf{A}$ y $\textbf{B}$ son iguales si son del mismo tamaño y sus componentes correspondientes son iguales $A_{i,j}=B_{i,j}$

### Matriz transpuesta
Es la matriz que resulta de intercambiar sus filas por sus columnas

Si

$$
\textit{A} = \begin{pmatrix}
A_{0,0} & A_{0,1} & \ldots & A_{0, n-1} \\
A_{1,0} & A_{1,1} & \ldots & A_{1, n-1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} & A_{m-1,1} & \ldots & A_{m-1, n-1}
\end{pmatrix}
$$

entonces,

$$
A^T = \begin{pmatrix}
A_{0,0} & A_{1,0} & \ldots & A_{m-1,0} \\
A_{0,1} & A_{1,1} & \ldots & A_{m-1,1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{0,n-1} & A_{1,n-1} & \ldots & A_{m-1,n-1}
\end{pmatrix}
$$

${A^T}_{i,j}=A_{j,i}$


**Ejemplo:**
<b>$\textit{B}$</b> = $ \begin{pmatrix}
1 & 4 &  2 & 0 & -1 \\
3 & 0 & -2 & 1 &  5 \\
\end{pmatrix}  $
entonces,
<b>$B^T$<b> = $ \begin{pmatrix}
1 & 3  \\
4 & 0  \\
2 & -2  \\
0 & 1  \\
-1 & 5  \\
\end{pmatrix}  $

### Suma de matrices

Dos matrices pueden sumarse si tienen el mismo tamaño:

$$\textit{C} = \textit{A} + \textit{B}$$

$$
\textit{C} = 
\begin{pmatrix}
A_{0,0} + B_{0,0} & A_{0,1} + B_{0,1} & \ldots & A_{0, n-1} + B_{0, n-1} \\
A_{1,0} + B_{1,0} & A_{1,1} + B_{1,1} & \ldots & A_{1, n-1} + B_{1, n-1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} + B_{m-1,0} & A_{m-1,1} + B_{m-1,1} & \ldots & A_{m-1, n-1} + B_{m-1, n-1}
\end{pmatrix}
$$

donde

$$C_{i,j} = A_{i,j} + B_{i,j}$$

### Suma de matrices con una escalar

La suma de una matriz por un escalar se realiza desarrollando la operación suma en cada uno de los elementos de la matriz

$ B = A + c $ donde $ B_{i,j} = A_{i,j} + c $

$\textit{B}$ = $ \begin{pmatrix}
A_{0,0} & A_{0,1} & \ldots & A_{0,n-1} \\
A_{1,0} & A_{1,1} & \ldots & A_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} & A_{m-1,1} & \ldots & A_{m-1,n-1}
\end{pmatrix} + c = $
$ \begin{pmatrix}
A_{0,0} + c & A_{0,1} + c & \ldots & A_{0,n-1} + c \\
A_{1,0} + c & A_{1,1} + c & \ldots & A_{1,n-1} + c \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} + c & A_{m-1,1} + c & \ldots & A_{m-1,n-1} + c
\end{pmatrix}$

### Producto de una matriz con un escalar

El producto de una matriz por un escalar se realiza desarrollando la operación producto en cada uno de los elementos de la matriz

$ B = cA $ donde $ B_{i,j} = cA_{i,j} $

$B = c\textit{A}$ = $ c\begin{pmatrix}
A_{0,0} & A_{0,1} & \ldots & A_{0,n-1} \\
A_{1,0} & A_{1,1} & \ldots & A_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} & A_{m-1,1} & \ldots & A_{m-1,n-1}
\end{pmatrix} = $
$ \begin{pmatrix}
cA_{0,0} & cA_{0,1} & \ldots & cA_{0,n-1} \\
cA_{1,0} & cA_{1,1} & \ldots & cA_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
cA_{m-1,0} & cA_{m-1,1} & \ldots & cA_{m-1,n-1}
\end{pmatrix}$


### Suma de una matriz con un vector

La suma de una matriz con un vector no está definida en el álgebra lineal convencional, ya que sus dimensiones no coinciden.

Sin embargo, en algunos contextos (por ejemplo, en programación con ciertas bibliotecas de cálculo numérico), puedes encontrar operaciones que parecen sumar una matriz con un vector. Esto se suele hacer mediante una operación llamada "broadcasting."

$ C = A + b $ donde $ C_{i,j} = A_{i,j} + b_{j} $

$\textit{C}$ = $ \begin{pmatrix}
A_{0,0} & A_{0,1} & \ldots & A_{0,n-1} \\
A_{1,0} & A_{1,1} & \ldots & A_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} & A_{m-1,1} & \ldots & A_{m-1,n-1}
\end{pmatrix} + $
$ \begin{pmatrix}
b_{0} & b_{0} & \ldots & b_{0}
\end{pmatrix}$

O

$\textit{C}$ = $ \begin{pmatrix}
A_{0,0} & A_{0,1} & \ldots & A_{0,n-1} \\
A_{1,0} & A_{1,1} & \ldots & A_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} & A_{m-1,1} & \ldots & A_{m-1,n-1}
\end{pmatrix} + $
$ \begin{pmatrix}
b_{0} \\
b_{1} \\
\vdots \\
b_{m-1}
\end{pmatrix}$

### Multiplicación de matrices

El producto $𝑨𝑩$ se puede realizar si el número de columnas de $𝑨$ es el mismo que el
número de filas de $𝑩$.  
Si $𝑨$ es de la forma $𝑚$ x $𝑛$ y $B$ es de la forma $𝑛$ x $𝑝$, $𝑪 = 𝑨𝑩$ es de la
forma $𝑚$ x $p$

$  C_{i,j}= {\sum_{k}} A_{i,k}B_{k,j}  $
  
El elemento $  C_{i,j} $ se obtiene de multiplicar cada uno de los elementos de la fila $i$ de la matriz $A$ por el correspondiente elemento de la columna $j$ de la matriz $B$, y calcular la suma de todos esos productos.



  
__Ejemplo:__

$C = \begin{pmatrix}
2 & 0 & -3 \\
4 & 1 & 5 \\
\end{pmatrix}$
$\begin{pmatrix}
7 & -1 & 4 & 7 \\
2 & 5 & 0 & -4 \\
-3 & 1 & 2 & 3\\
\end{pmatrix}$
$ = \begin{pmatrix}
23 & -5 & 2 & 5 \\
15 & 6 & 26 & 39 \\
\end{pmatrix}$





- $C_{0,0} = A_{0,0}B_{0,0} + A_{0,1}B_{1,0} + A_{0,2}B_{2,0} = (2)(7) + (0)(2) + (-3)(-3) = 23$
- $C_{0,1} = A_{0,0}B_{0,1} + A_{0,1}B_{1,1} + A_{0,2}B_{2,1} = (2)(-1) + (0)(5) + (-3)(1) = -5$
- $C_{0,2} = A_{0,0}B_{0,2} + A_{0,1}B_{1,2} + A_{0,2}B_{2,2} = (2)(4) + (0)(0) + (-3)(2) = 2$
- $C_{0,3} = A_{0,0}B_{0,3} + A_{0,1}B_{1,3} + A_{0,2}B_{2,3} = (2)(7) + (0)(-4) + (-3)(3) = 5$

- $C_{1,0} = A_{1,0}B_{0,0} + A_{1,1}B_{1,0} + A_{1,2}B_{2,0} = (4)(7) + (1)(2) + (5)(-3) = 15$
- $C_{1,1} = A_{1,0}B_{0,1} + A_{1,1}B_{1,1} + A_{1,2}B_{2,1} = (4)(-1) + (1)(5) + (5)(1) = 6$
- $C_{1,2} = A_{1,0}B_{0,2} + A_{1,1}B_{1,2} + A_{1,2}B_{2,2} = (4)(4) + (1)(0) + (5)(2) = 26$
- $C_{1,3} = A_{1,0}B_{0,3} + A_{1,1}B_{1,3} + A_{1,2}B_{2,3} = (4)(7) + (1)(-4) + (5)(3) = 39$

### Producto punto

El producto punto de dos vectores $x$ y $y$ es el producto de matrices $x^T y$

Si $x = \begin{pmatrix}
x_0  \\
x_1 \\
\vdots \\
x_{n-1} \\
\end{pmatrix}$
$\;$ y $y = \begin{pmatrix}
y_0  \\
y_1 \\
\vdots \\
y_{n-1} \\
\end{pmatrix}$
, $\;$ $x^T y =  \begin{pmatrix}
x_0  & x_1 & \ldots & x_{n-1}  \\
\end{pmatrix}$
$\begin{pmatrix}
y_0  \\
y_1 \\
\vdots \\
y_{n-1} \\
\end{pmatrix}$ $=$
$x_{0}y_{0} + x_{1} y_{1} + \ldots + x_{n-1} y_{n-1}$



### Producto Hadamard

Es un producto directo que se obtiene al multiplicar elemento por elemento de dos matrices

$𝐶 = 𝐴⊙𝐵$

donde $𝐶_{𝑖𝑗} = 𝐴_{𝑖,𝑗}𝐵_{𝑖,𝑗}$

$\textit{C}$ = $ \begin{pmatrix}
A_{0,0} & A_{0,1} & \ldots & A_{0,n-1} \\
A_{1,0} & A_{1,1} & \ldots & A_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} & A_{m-1,1} & \ldots & A_{m-1,n-1}
\end{pmatrix} ⊙  \begin{pmatrix}
B_{0,0} & B_{0,1} & \ldots & B_{0,n-1} \\
B_{1,0} & B_{1,1} & \ldots & B_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
B_{m-1,0} & B_{m-1,1} & \ldots & B_{m-1,n-1}
\end{pmatrix}$

### División

En el contexto del álgebra lineal, la división de matrices como tal no está definida. Sin embargo, en NumPy, puedes realizar una división elemento por elemento (al igual que la multiplicación de Hadamard) usando el operador /. Esto dividirá cada elemento en la matriz de la izquierda por el elemento correspondiente en la matriz de la derecha.

$\textit{C}$ = $ \begin{pmatrix}
A_{0,0} & A_{0,1} & \ldots & A_{0,n-1} \\
A_{1,0} & A_{1,1} & \ldots & A_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m-1,0} & A_{m-1,1} & \ldots & A_{m-1,n-1}
\end{pmatrix} /  \begin{pmatrix}
B_{0,0} & B_{0,1} & \ldots & B_{0,n-1} \\
B_{1,0} & B_{1,1} & \ldots & B_{1,n-1} \\
\vdots & \vdots & \ddots & \vdots \\
B_{m-1,0} & B_{m-1,1} & \ldots & B_{m-1,n-1}
\end{pmatrix}$


```python
B = np.array([[7,-1,4,7],[2,5,0,-4],[-3,1,2,3]])
B / B
```
